### Viola-Jones Face Detection

```python
import cv2
from cv2 import CascadeClassifier
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = plt.imread('Images/photo.jpg')
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'xticks':(), 'yticks':()})
ax.imshow(image)

model = CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# use the minNeighbors param to mmake the classifier more selective
faces = model.detectMultiScale(image, minNeighbors=3, minSize=(250, 250))

for face in faces:
    x, y, w, h = face
    rect = Rectangle((x, y), w, h, fill=False, color='red', lw=2)
    ax.add_patch(rect)
    
```
